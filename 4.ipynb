```python
import numpy as np
#create 1D array contain number fron 1 to 24
print("1D Array:")
arr=np.arange(1,25)   #using np.arange()
print(arr)
#resape it into (4,6)
v1=arr.reshape(4,6)
print("Reshape into (4x6):","\n",v1)
#reshape it into (3,8)
v2=arr.reshape(3,8)
print("Reshape into (3x8):","\n",v2)
#reshape it into(2,3,4)(3D array)
v3=arr.reshape(2,3,4)
print("Reshape into (2,3,4)","\n",v3)
```
